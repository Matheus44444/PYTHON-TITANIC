```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import random
np.random.seed(42)
random.seed(42)

# Configurações de estilo
sns.set(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (8, 5)

# Carrega o dataset Titanic
data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
data.head()
```

```python
sns.countplot(data=data, x="sex", hue="survived")
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x="age", hue="survived", kde=True, bins=30)
plt.title("Distribuição de Idades por Sobrevivência")
plt.show()

```

```python
sns.histplot(data=data, x="age", hue="survived", multiple="stack", bins=20)
sns.barplot(data=data, x="pclass", y="survived", hue="sex")
plt.title("Taxa de Sobrevivência por Classe e Gênero")
plt.show()

```

```python
sns.barplot(data=data, x="class", y="survived", estimator=lambda x: sum(x)/len(x))
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre Variáveis Numéricas")
plt.show()
```

```python
sns.barplot(data=data, x="class", y="survived", estimator=lambda x: sum(x)/len(x))
plt.title("Taxa Média de Sobrevivência por Classe")
plt.xlabel("Classe do Bilhete")
plt.ylabel("Proporção de Sobreviventes")
plt.show()
```

```python
corr = data.select_dtypes(include=["number"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlação das Variáveis Numéricas")
plt.show()
```

```python
sns.catplot(data=data, x="embarked", hue="survived", col="class",
            kind="count", height=4, aspect=0.8)
plt.subplots_adjust(top=0.8)
plt.suptitle("Sobrevivência por Porto de Embarque e Classe")
plt.show()
```
